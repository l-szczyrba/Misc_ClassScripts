```python
df_money2 = df_money.groupby(['Date']).sum()
df_money2.reset_index(inplace=True)
df_money2
```

```python
plt.figure(figsize=(15,7))
fig2 = sns.lineplot(x=df_money2.Date, y=df_money2.Amount,
             data=df_money2, err_style= None)

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
fig2.yaxis.set_major_formatter(tick) 

fig2.set_title('Money Raised Throughout October', size = 23)
fig2.set_ylabel('Dollar Amount', size = 18)
fig2.set_xlabel('', size = 18)
fig2.tick_params(labelsize=15)
plt.savefig('Money.png', dpi = 300, pad_inches = 1)
```
